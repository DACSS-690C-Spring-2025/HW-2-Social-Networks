```python
import networkx as nx
import pandas as pd

# Read data from GitHub
PeruFamilyData = "https://github.com/DACSS-CSSmeths/Networks_intro/raw/refs/heads/main/casesData/dataFigueroa.xlsx"

# Load adjacency matrix
FigueData = pd.read_excel(PeruFamilyData, sheet_name='Relationships', index_col=0)

# Load node attributes
FigueAttr = pd.read_excel(PeruFamilyData, sheet_name='Attributes')

# Create undirected graph
G_figue = nx.from_pandas_adjacency(FigueData)

# Create dictionary for node attributes
attr_dict = {row['node']: {'Multinacional': row['Multinacional']} for _, row in FigueAttr.iterrows()}

# Assign attributes to nodes
nx.set_node_attributes(G_figue, attr_dict)

# Save graph as .graphml file for visualization
graphml_filename = "figueroa_network.graphml"
nx.write_graphml(G_figue, graphml_filename)

print(f"GraphML file saved as {graphml_filename}")
```

```python
import matplotlib.pyplot as plt
# Load graph

G = nx.read_graphml("figueroa_network.graphml")

# Draw the network
plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=True, node_color="lightblue", edge_color="gray", font_size=8)
plt.show()
```

```python
# Spread out nodes more for readability
G = nx.read_graphml("figueroa_network.graphml")
pos = nx.spring_layout(G, k=0.8)

# Draw the network
plt.figure(figsize=(12, 10))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_color="lightblue",
    edge_color="gray",
    font_size=10,
    node_size=800,
    font_color="black"
)

plt.title("Figueroa Family Network")
plt.show()
```
